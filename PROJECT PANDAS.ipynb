```python
import pandas as pd


students = {
    "Name": [
        "Ahmad", "Sara", "Mustafa",
        "Nilofar", "Pari", "Alix",
        "Sina", "Ershad", "Marwa"
    ],
    "Class": [
        "A", "A", "B",
        "B", "C", "C",
        "A", "B", "C"
    ],
    "Grade": [
        87, 89, 76,
        90, 87, 59,
        98, 90, 76
    ]
}

df = pd.DataFrame(students)

print("========== STUDENT DATA ==========")
print(df)

print("\n========== SUMMARY STATISTICS ==========")
print(df.describe())

print("\n========== AVERAGE GRADE ==========")
print(df.groupby("Name")["Grade"].mean())

print("\n========== HIGHEST GRADE ==========")
print(df.groupby("Name")["Grade"].max())

print("\n========== LOWEST GRADE ==========")
print(df.groupby("Name")["Grade"].min())

print("\n========== COMPLETE ANALYSIS ==========")
analysis = df.groupby("Name")["Grade"].agg([
    "max",
    "min",
    "mean",
    "count"
])

print(analysis)

best_student = df.loc[df["Grade"].idxmax()]

print("\n========== TOP STUDENT ==========")
print(f"Name : {best_student['Name']}")
print(f"Class: {best_student['Class']}")
print(f"Grade: {best_student['Grade']}")
```
